### Name

```python
df.filter(pl.col("Name").is_null())
```

- Column name should contains name of the business that applied for the loan

- It's hard to identify whether these missing values are MNAR or MAR because there are only 3 data points.
- Hece I will tag these rows with a new column

```python
loan_ids_with_missing_name = df.filter(pl.col("Name").is_null())["LoanNr_ChkDgt"].to_list()
loan_ids_with_missing_name
```

```python
df = df.with_columns(
        pl.when(pl.col("Name").is_null())
        .then(True)
        .otherwise(None)
        .alias("needs_to_verify")
    ).with_columns(
        pl.when(pl.col("LoanNr_ChkDgt").is_in(loan_ids_with_missing_name))
        .then(["missing value on name"])
        .otherwise([])
        .alias("verify_reasons")
    )

display(df.filter(pl.col("verify_reasons").list.len().gt(0)))

del loan_ids_with_missing_name
```

### Zip

```python
df.filter(pl.col("Zip").cast(pl.String).str.len_chars().le(4)).tail(1)
```

```python
df.filter(pl.col("Zip").cast(pl.String).str.len_chars().gt(5))
```

- US's zip code has to consist of 5 digits
- There are some samples of zip codes in the dataset, ex: 6811, 2816 correctly refer to CT and RI state respectfully. These zip codes should be 06811 and 02816

```python
df = df.cast({"Zip": pl.String})
```

```python
df = df.with_columns(
    pl.when(pl.col("Zip").str.len_chars().eq(4))
    .then(pl.lit("0") + pl.col("Zip"))
    .otherwise("Zip")
    .alias("Zip")
)

df.filter(pl.col("Zip").str.starts_with("0")).tail(5)
```

```python
zip_code_le_3_digits = df.filter(pl.col("Zip").cast(pl.String).str.len_chars().le(3))["Zip"].to_list()
len(zip_code_le_3_digits)
```

```python
df = df.with_columns(
    pl.when(pl.col("Zip").is_in(zip_code_le_3_digits))
        .then(True)
        .otherwise("needs_to_verify")
        .alias("needs_to_verify"),
    pl.when(pl.col("Zip").is_in(zip_code_le_3_digits))
        .then(
            pl.concat_list(
                pl.col("verify_reasons"),
                pl.lit("invalid_zip_code")
            )
        )
        .otherwise("verify_reasons")
        .alias("verify_reasons")
)
```

```python
zip_code_le_3_digits[40: 50]
```

- There are zip codes with char length <= 3 digits. These zip codes are invalid
- I will change these zip codes into unknown

```python
df = df.with_columns(
    pl.when(pl.col("Zip").is_in(zip_code_le_3_digits))
    .then(pl.lit("invalid"))
    .otherwise("Zip")
    .name.keep()
)

del zip_code_le_3_digits
```

### City

> This column describe borrower's origin city

- America has some cities that consist of 3 digit names, such as: rye, lee, jal
- I will inspect whether or not cities with 2 digit names or less exist

```python
df.filter(pl.col("City").str.len_chars().le(2))['City']
```

> There are 86 cities with only 1 to 2 digit characters

```python
df.filter(pl.col("City").is_null()).shape
```

> The dataset also has 30 missing values of city

```python
df = df.with_columns(
    pl.when((pl.col("City").str.len_chars().le(2)) | (pl.col("City").is_null()))
        .then(pl.lit("invalid"))
        .otherwise("City")
        .name.keep(),
    pl.when((pl.col("City").str.len_chars().le(2)) | (pl.col("City").is_null()))
        .then(True)
        .otherwise("needs_to_verify")
        .alias("needs_to_verify"),
    pl.when((pl.col("City").str.len_chars().le(2)) | (pl.col("City").is_null()))
        .then(
            pl.concat_list(
                pl.col("verify_reasons"),
                pl.lit("invalid city name")
            )
        )
        .otherwise("verify_reasons")
        .alias("verify_reasons")
)
```
